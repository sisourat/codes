## Particle Under Harmonic Potential

The aim of this small project is to implement Verlet algorithm to numerically solve the Newton's equation of motion for a particle of mass $m$ under a harmonic potential $V(x)=\frac{1}{2}m \omega ^2 x^2$ in one space dimension. Using the fact that $\vec F = -\vec \nabla V$, the equation of our interest becomes $$ m \frac{d^2 x(t)}{dt^2} = -k x(t) $$ or equivalently $$ \frac{d^2 x(t)}{dt^2} = -\omega ^2 x(t) $$ where $k=m \omega^2$ is the force constant. The analytical solution of this equation is $$ x(t) = x(0)\cos(\omega t) $$



Now, we need to solve for $x(t)$ numerically by using Verlet algorithm. Consider the Taylor expansion of the following functions

$$x(t+\Delta t) = x(t) + \Delta t \frac{dx(t)}{dt} + \frac{\Delta t ^2}{2!}\frac{d^2x(t)}{dt^2}+\frac{\Delta t ^3}{3!}\frac{d^3x(t)}{dt^3} + \mathcal{O}(\Delta t ^4)$$

$$x(t-\Delta t) = x(t) - \Delta t \frac{dx(t)}{dt} + \frac{\Delta t ^2}{2!}\frac{d^2x(t)}{dt^2}-\frac{\Delta t ^3}{3!}\frac{d^3x(t)}{dt^3} + \mathcal{O}(\Delta t ^4)$$

By summing the two terms above, we obtain the following integrator to calculate the position in the next time step.

$$x(t+\Delta t) = 2x(t) - x(t - \Delta t) + \Delta t^2 \frac{d^2x(t)}{dt^2} + \mathcal{O}(\Delta t ^4)$$


In order to use this integrator, we need the values of $x(t)$ and $x(t - \Delta t)$ to start the iteration. Since we are trying to solve a second order differential equation, we need two initial conditions. In the following, we write a function for Verlet algorithm which asks for these two initial conditions (namely $x(0)$ and $v(0)$) and then iterate to solve the equation. At the end we will compare the numerical and the analytical results.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def Verlet (x_init,v_init,m,k,time_step,sim_time):
    """
    This function takes 6 inputs: Initial position (x_init), initial velocity (v_init), mass (m) of 
    the particle, force constant (k), time step, and the simulation time (sim_time) respectively.
    
    It returns 2 lists: The position (x) of the particle and the corresponding time interval (t)
    """
    
    x = [x_init]             # initialise the position 
    v = [v_init]             # initialise the velocity 
    t = [0]                  # initialise the time 
    dt = time_step
    global time
    time = sim_time
    global omega
    omega = (k/m)**0.5       # define omega
    
    i = 0                    # initialise iteration variable 
    
    for i in range(int(time/dt)+1):                    # iterate until the desired simulation time is reached 
        
        a = -(omega**2) * x[i]                       # define acceleration 
        
        if i == 0:            # we need at least two values of x to use the integrator
                              # this if condition generates the second value by using a basic Taylor approximation
            
            x_inc = x[i] + dt*v[i] + 0.5*(dt)**2 * a           # Taylor expansion
            
        else:             # use Verlet integrator
            
            x_inc = 2*x[i] - x[i-1] + (dt)**2 * a
            
        x.append(x_inc)     # append the incremented value of x to the existing list
        t.append(t[i]+dt)   # append the incremented value of t to the existing list
        
    return(x,t)             # return position and time
```

```python
x,t = Verlet(1,0,1,1,0.1,20)      # do not set x=0 to observe the sinusoidal behaviour of the solution


x_an = x[0] * np.cos(omega * np.array(t))  # analytical solution
```

```python
# Compare the analytical and numerical results

plt.plot(t,x_an,'b',label="Analytical")
plt.plot(t,x,'ro',label="Numerical", markeredgewidth=1,
         markersize=4, markevery=3)
#plt.ylim(-1,1)
plt.xlim(0,time)
plt.title("Numerical and Analytical Solution")
plt.xlabel("Time")
plt.ylabel("Position")
plt.legend(bbox_to_anchor=(1.02, 1.0), loc='upper left')
plt.show()
```

We observe that the numerical solution that is obtained by using the Verlet algorithm matches well with the analytical solution.

```python
# Absolute error 

err = abs(x_an - x)

plt.errorbar(t,x,yerr=err,fmt='ko',markevery=3,markeredgewidth=1,label="Numerical",
             markersize=2,ecolor='r',elinewidth=2,capsize=4,errorevery=3)
plt.title("Absolute Error")
plt.xlabel("Time")
plt.ylabel("Position")
plt.legend(bbox_to_anchor=(1.02, 1.0), loc='upper left')
plt.show()
```

If we want to make a thorough comparison, we can calculate the absolute error and plot the corresponding error bars. In this case, we can clearly see that the error is small in the extrema where both the first and the second derivative are zero. Also, the error is maximised as the absolute value of the first derivative is increased.

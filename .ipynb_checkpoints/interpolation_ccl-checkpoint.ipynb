# Interpolation C-Cl

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate
%matplotlib inline
```

```python
def derive(V,x,i):
    """
    This function returns the first derivative of V with respect to x at a given point i
    """
    if i in range(len(x)):
        
        if i == 0:
            
            g = (V[1] - V[0]) / (x[1] - x[0])
            
        elif i == len(x) - 1:
            
            g = (V[-1] - V[-2]) / (x[-1] - x[-2])
            
        else:
            
            g = (V[i+1] - V[i-1]) / (x[i+1] - x[i-1])
    
    return g
```

```python
def Verlet (x,V,t0,tf,N):
    """
    This function takes the following inputs: 
    Position(x), potential(V), initial time(t0), final time(tf), number of steps(N)
    
    It returns the solution (xt) of the equation of motion on a time interval (t) in [t0,tf]
    """
    dt = (tf-t0)/N
    xt = np.zeros(int(N+1))
    t = np.zeros(int(N+1))
    m = (35.4530 * 12.0107) /(35.4530 + 12.0107)    # reduced mass of C-Cl
    xt[0] = x[0]
    t[0] = t0
    for i in range(N):                     
        
        a = -derive(V,x,i)/m                       # define acceleration 
        
        if i == 0:            # we need at least two values of x to use the integrator
            
            t[i+1] = t0 + dt
            xt[i+1] = xt[i] + 0.5*(dt)**2 * a               # Taylor expansion (assuming zero initial velocity)
            
            # if there is an initial velocity, uncomment the line below, and add v_init as an input
            
            #xt[i+1] = x[i] + dt*v_init + 0.5*(dt)**2 * a     
         
        
        else:             # use Verlet integrator
            
            t[i+1] = t[i] + dt
            xt[i+1] = 2*xt[i] - xt[i-1] + (dt)**2 * a
        
        
        
    return(t,xt)             # return the solution
```

## Interpolation

Read the data from a file and then evaluate the interpolating function in a desired interval.

```python
# the first column contains the position

x = np.loadtxt("myData.txt", delimiter=' ',skiprows=2)[:,0]       

# other columns contain energy values

E = [np.loadtxt("myData.txt", delimiter=' ',skiprows=2)[:,i] for i in range(1,8)]

```

```python
for i in range(len(E)): 
    
    plt.plot(x,E[i],label="E" + str(i+1))

plt.legend()
plt.xlabel("Position")
plt.ylabel("Energy")
plt.show()
```

```python
# Solution by Verlet algorithm       ( Verlet (x,V,t0,tf,dt) )



t_verlet = [Verlet(x,E[i],0,10,100)[0] for i in range(len(E))]
x_verlet = [Verlet(x,E[i],0,10,100)[1] for i in range(len(E))]
 
for i in range(len(E)):    
    plt.plot(t_verlet[i],x_verlet[i],label="E"+str(i+1))

plt.legend()
plt.xlabel("Time")
plt.ylabel("Position")
plt.show()
```

Now, we create interpolating functions for distinct energy values.

```python
# Create the interpolating function

f = [interpolate.interp1d(x,E[i],kind="cubic",fill_value="extrapolate") for i in range(len(E))]
```

```python
# Evaluate the interpolating function in a new interval (x_test)

x_test = np.linspace(2,5,len(x))

for i in range(len(E)):
    plt.plot(x_test,f[i](x_test),label="E"+str(i+1))
plt.legend(loc="upper right")
plt.xlabel("Position")
plt.ylabel("Energy")
plt.show()
```

```python
# Solution by Verlet algorithm in the new interval

t2_verlet = [Verlet(x_test,f[i](x_test),0,10,100)[0] for i in range(len(E))]
x2_verlet = [Verlet(x_test,f[i](x_test),0,10,100)[1] for i in range(len(E))]


for i in range(len(E)):    
    plt.plot(t2_verlet[i],x2_verlet[i],label="E"+str(i+1))

plt.legend()
plt.xlabel("Time")
plt.ylabel("Position")
plt.show()
```

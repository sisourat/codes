# Interpolation C-Cl

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate
%matplotlib inline
```

```python
def derive(V,x,i):
    """
    This function returns the first derivative of V with respect to x at a given point i
    """
    if i in range(len(x)):
        
        if i == 0:
            
            g = (V[1] - V[0]) / (x[1] - x[0])
            
        elif i == len(x) - 1:
            
            g = (V[-1] - V[-2]) / (x[-1] - x[-2])
            
        else:
            
            g = (V[i+1] - V[i-1]) / (x[i+1] - x[i-1])
    
    return g
```

```python
def Verlet (x,V,time_step):
    """
    This function takes 3 inputs: Position(x), potential(V), and the time step
    
    It returns the solution (xt) of the equation of motion for a particle of mass m
    """
    
    dt = time_step
    i = 0                    # initialise iteration variable 
    xt = []
    m = (35.4530 * 12.0107) /(35.4530 + 12.0107)    # reduced mass of C-Cl
    
    for i in range(len(x)):                     
        
        a = -derive(V,x,i)/m                       # define acceleration 
        
        if i == 0:            # we need at least two values of x to use the integrator
            
            x_inc = x[i] + 0.5*(dt)**2 * a               # Taylor expansion (assuming zero initial velocity)
            
            # if there is an initial velocity, uncomment the line below, and add v_init as an input
            
            #x_inc = x[i] + dt*v_init + 0.5*(dt)**2 * a     
         
        
        else:             # use Verlet integrator
            
            x_inc = 2*x[i] - x[i-1] + (dt)**2 * a
            
        xt.append(x_inc)     # append the incremented value of x to the existing list
        
        
    return(xt)             # return the solution
```

## Interpolation

Read the data from a file and then evaluate the interpolating function in a desired interval.

```python
# the first column contains the position

x = np.loadtxt("myData.txt", delimiter=' ',skiprows=2)[:,0]       

# other columns contain energy values

E = [np.loadtxt("myData.txt", delimiter=' ',skiprows=2)[:,i] for i in range(1,8)]

```

```python
for i in range(len(E)): 
    
    plt.plot(x,E[i],label="E" + str(i+1))

plt.legend()
plt.xlabel("Position")
plt.ylabel("Energy")
plt.show()
```

```python
# Solution by Verlet algorithm

dt = 0.01  # time step

x_verlet = [Verlet(x,E[i],dt) for i in range(len(E))]

t = np.linspace(0,dt*len(x),len(x))   # plot the position in this time interval


for i in range(len(E)): 
    
    plt.plot(t,x_verlet[i],label="E" + str(i+1))

plt.legend()
plt.xlabel("Time")
plt.ylabel("Position")
plt.show()
```

Now, we create interpolating functions for distinct energy values.

```python
# Create the interpolating function

f = [interpolate.interp1d(x,E[i],kind="cubic",fill_value="extrapolate") for i in range(len(E))]
```

```python
# Evaluate the interpolating function in a new interval (x_test)

x_test = np.linspace(2,5,len(x))

for i in range(len(E)):
    plt.plot(x_test,f[i](x_test),label="E"+str(i+1))
plt.legend(loc="upper right")
plt.xlabel("Position")
plt.ylabel("Energy")
plt.show()
```

```python
# Solution by Verlet algorithm in the new interval


x_test_verlet = [Verlet(x_test,f[i](x_test),dt) for i in range(len(E))]

for i in range(len(E)): 
    plt.plot(t,x_test_verlet[i],label="E" + str(i+1))
plt.legend()
plt.xlabel("Time")
plt.ylabel("Position")
plt.show()
```
